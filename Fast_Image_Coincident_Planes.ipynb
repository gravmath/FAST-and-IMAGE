```python
import datetime as dt
import pickle
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Ffile = 'C:/Users/Conrad/Documents/GitHub/FAST-and-Image/FAST_states_pickled.p'
Ifile = 'C:/Users/Conrad/Documents/GitHub/FAST-and-Image/IMAGE_states_pickled.p'
FEpochs, FStates, FL = pickle.load(open(Ffile,'r'))
IEpochs, IStates, IL = pickle.load(open(Ifile,'r'))
```

```python
print 'FAST Epoch Span:  ', FEpochs[0], FEpochs[-1]
print 'IMAGE Epoch Span: ', IEpochs[0], IEpochs[-1]
```

```python
dots   = []
epochs = []
fcounter = 0
for fepoch in FEpochs:
    icounter = 0
    for iepoch in IEpochs:
        if abs((fepoch - iepoch).days) < 1:
            print fcounter, icounter, fepoch, iepoch
            fl_mag = np.sqrt(FL[fcounter].dot(FL[fcounter]))
            il_mag = np.sqrt(IL[icounter].dot(IL[icounter]))            
            dot    = FL[fcounter].dot(IL[icounter])/fl_mag/il_mag
            dots.append(dot)
            epochs.append(fepoch)
        icounter = icounter + 1
    fcounter = fcounter + 1
```

```python
plt.plot(epochs,np.arccos(dots)*180/np.pi)
plt.xlabel('FAST-IMAGE Epoch')
plt.ylabel('Plane Alignment (deg)')
plt.grid('on')
```
